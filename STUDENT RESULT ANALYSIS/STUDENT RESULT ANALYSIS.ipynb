DATA VISUALIZATION

DESCRIBING THE GENDER DISTRIBUTION AMONG STUDENTS.

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x='Gender')
ax.bar_label(ax.containers[0])
plt.title('Gender Distribution', color = 'blue',fontsize=15)
plt.show()
```

FROM THE ABOVE BAR GRAPH WE CAN CONCLUDE THAT THE NUMBER OF FEMALES IN THE DATA IS MORE THAN THE MALES.

ANALYSING THE IMPACT OF PARENT'S EDUCATION ON STUDENT'S SCORE

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
gb
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.title("Impact of Parents's education on Student's Score")
plt.show()
```

FROM THE ABOVE HEATMAP WE CAN CONCLUDE THAT THE EDUCATION OF THE PARENTS HAVE A GOOD IMPACT ON THEIR CHILDREN'S SCORES.

 THE IMPACT OF PARENTAL MARITAL STATUS ON STUDENTS' SCORES.

```python
gb1=df.groupby('ParentMaritalStatus').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
gb1
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot=True)
plt.title("Relationship between Parent's Martial Status and Student's Score")
plt.show()
```

FROM THE ABOVE HEATMAP WE CAN CONCLUDE THAT THE MARITAL STATUS OF PARENTS DOES NOT PLAY A SIGNIFICANT ROLE IN DETERMINING STUDENTS' SUCCESS IN EXAMS.

DISPLAYING DISTRIBUTION OF ETHINIC GROUPS

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["group A", "group B", "group C", "group D", "group E"]
myList = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
print('myList is:', myList)
```

```python
plt.pie(myList, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

FROM THE ABOVE PIE CHART WE CAN CONLUDE THAT GROUP C IS THE LARGEST AND GROUP A IS THE SMALLEST

ANALYZING THE NUMBER OF STUDENTS ACROSS DIFFERENT ETHNIC GROUPS

```python
ax = sns.countplot(data=df, x="EthnicGroup", hue="EthnicGroup", palette="Set2", legend=False)
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

FROM THE ABOVE BAR GRAPH WE CAN CONCLUDE THAT GROUP C (ETHINICITY) CONTAINS MAXIMUM NUMBER OF STUDENTS AND GROUP A (ETHINICITY) CONTAINS LEAST NUMBER OF STUDENTS
